```python
from googleapiclient.discovery import build
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Ensure VADER lexicon is available
nltk.download('vader_lexicon')

# Use your API key here
API_KEY = "AIzaSyAr0AVCWi0LBj-YxTqjTUR3smejRQhVF6s"
VIDEO_ID = "bD5Zd0KWLqA"  # Extracted from the YouTube video URL

# Function to extract comments from YouTube
def get_youtube_comments(api_key, video_id, max_results=50):
    youtube = build('youtube', 'v3', developerKey=api_key)

    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=max_results,
        textFormat="plainText"
    )

    response = request.execute()

    for item in response.get("items", []):
        comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
        comments.append(comment)

    return comments

# Function to perform sentiment analysis
def analyze_sentiment(comments):
    sia = SentimentIntensityAnalyzer()
    sentiment_results = []

    for comment in comments:
        sentiment_score = sia.polarity_scores(comment)  # Get sentiment scores
        sentiment = (
            "Positive" if sentiment_score["compound"] > 0.05 else
            "Negative" if sentiment_score["compound"] < -0.05 else
            "Neutral"
        )
        sentiment_results.append((comment, sentiment_score["compound"], sentiment))

    return sentiment_results

if __name__ == "__main__":
    try:
        # Extract comments
        comments = get_youtube_comments(API_KEY, VIDEO_ID)

        # Perform sentiment analysis
        sentiment_results = analyze_sentiment(comments)

        # Print results
        for idx, (comment, score, sentiment) in enumerate(sentiment_results, start=1):
            print(f"{idx}. {comment}\n   Sentiment: {sentiment} (Score: {score})\n")

    except Exception as e:
        print(f"Error: {e}")
```

```python
import matplotlib.pyplot as plt
from googleapiclient.discovery import build
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

# Download VADER Lexicon
nltk.download('vader_lexicon')

# Use your API key here
API_KEY = "AIzaSyAr0AVCWi0LBj-YxTqjTUR3smejRQhVF6s"
VIDEO_ID = "bD5Zd0KWLqA"  # Extracted from YouTube URL

# Function to extract comments
def get_youtube_comments(api_key, video_id, max_results=50):
    youtube = build('youtube', 'v3', developerKey=api_key)

    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=max_results,
        textFormat="plainText"
    )

    response = request.execute()

    for item in response.get("items", []):
        comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
        comments.append(comment)

    return comments

# Function to perform sentiment analysis
def analyze_sentiment(comments):
    sia = SentimentIntensityAnalyzer()

    sentiment_results = []
    sentiment_counts = {"Positive": 0, "Negative": 0, "Neutral": 0}

    for comment in comments:
        sentiment_score = sia.polarity_scores(comment)
        sentiment = (
            "Positive" if sentiment_score["compound"] > 0.05 else
            "Negative" if sentiment_score["compound"] < -0.05 else
            "Neutral"
        )

        sentiment_results.append((comment, sentiment_score["compound"], sentiment))
        sentiment_counts[sentiment] += 1

    return sentiment_results, sentiment_counts

# Function to plot sentiment distribution
def plot_sentiment(sentiment_counts):
    labels = list(sentiment_counts.keys())
    values = list(sentiment_counts.values())

    plt.figure(figsize=(6, 4))
    plt.bar(labels, values, color=['green', 'red', 'blue'])

    plt.xlabel("Sentiment")
    plt.ylabel("Count")
    plt.title("Sentiment Analysis of YouTube Comments")
    plt.show()

if __name__ == "__main__":
    try:
        # Extract comments
        comments = get_youtube_comments(API_KEY, VIDEO_ID)

        # Perform sentiment analysis
        sentiment_results, sentiment_counts = analyze_sentiment(comments)

        # Print results
        for idx, (comment, score, sentiment) in enumerate(sentiment_results, start=1):
            print(f"{idx}. {comment}\n   Sentiment: {sentiment} (Score: {score})\n")

        # Plot sentiment distribution
        plot_sentiment(sentiment_counts)

    except Exception as e:
        print(f"Error: {e}")
```

```python
import nltk
import matplotlib.pyplot as plt
from googleapiclient.discovery import build
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required NLTK model
nltk.download('vader_lexicon')

# üîë Replace with your YouTube API Key
API_KEY = "AIzaSyAr0AVCWi0LBj-YxTqjTUR3smejRQhVF6s"
VIDEO_ID = "bD5Zd0KWLqA"  # Extract only the video ID, not full URL

# Function to fetch comments from YouTube
def get_youtube_comments(api_key, video_id, max_results=50):
    youtube = build('youtube', 'v3', developerKey=api_key)
    comments = []

    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=max_results,
        textFormat="plainText"
    )

    response = request.execute()

    for item in response.get("items", []):
        comment = item["snippet"]["topLevelComment"]["snippet"]["textDisplay"]
        comments.append(comment)

    print(f"\n‚úÖ Total Comments Fetched: {len(comments)}")
    return comments

# Function to analyze sentiment
def analyze_sentiment(comments):
    sia = SentimentIntensityAnalyzer()

    sentiment_counts = {"Positive": 0, "Negative": 0, "Neutral": 0}

    for comment in comments:
        sentiment_score = sia.polarity_scores(comment)["compound"]

        if sentiment_score > 0.05:
            sentiment_counts["Positive"] += 1
        elif sentiment_score < -0.05:
            sentiment_counts["Negative"] += 1
        else:
            sentiment_counts["Neutral"] += 1

    return sentiment_counts

# Function to plot sentiment distribution
def plot_sentiment_distribution(sentiment_counts):
    labels = list(sentiment_counts.keys())
    values = list(sentiment_counts.values())

    plt.figure(figsize=(6, 4))
    plt.bar(labels, values, color=["green", "red", "gray"])
    plt.xlabel("Sentiment")
    plt.ylabel("Number of Comments")
    plt.title("YouTube Comments Sentiment Analysis")
    plt.show()

# Main Execution
if __name__ == "__main__":
    comments = get_youtube_comments(API_KEY, VIDEO_ID)
    sentiment_counts = analyze_sentiment(comments)
    print(f"\nüìä Sentiment Counts: {sentiment_counts}")
    plot_sentiment_distribution(sentiment_counts)
```
